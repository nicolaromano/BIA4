```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
img = plt.imread("cells.jpg")

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

x0, x1, y0, y1 = (605, 635, 420, 450)

ax[0].imshow(img)
ax[0].axis("off")
r = patches.Rectangle((x0, y0), x1-x0, y1-y0, linewidth=2, facecolor="none", edgecolor="red")
ax[0].add_patch(r)

ax[1].imshow(img[y0:y1, x0:x1])
ax[1].axis("off")

for i in range(30):
    ax[1].axline((i-0.5, 0), (i-0.5, 25), linewidth = 0.5, color="r")
    ax[1].axline((0, i-0.5), (25, i-0.5), linewidth=0.5, color="r")
```
