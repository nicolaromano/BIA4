## How do I import Matplotlib in Python?

You usually import the pyplot module from Matplotlib using:

```python
import matplotlib.pyplot as plt
```

## How do I plot a simple line chart?

Load the data you want to plot and use the `plot` function from Matplotlib.

Remember to use `show` to display the plot.

In the following example, we plot the function $y = x^2$ for $x = [0, 1, \dots, 10]$ (obviously, you can read your data from a file or plot the result of some computation in the same way).



```python
x = range(10)
# We use a list comprehension to generate the y values
# This generates a list (because of the square brackets) of the squares 
# of the values in x (using the ** operator for exponentiation)
squares = [x_value**2 for x_value in x]
plt.plot(x, squares)
plt.show()
```

## How do I create multiple plots in the same figure?

You can use the `subplots` function to create multiple plots in the same figure.

The `subplots` function returns a figure and an array of axes, which you can use to plot your data.

For example, here is a figure with 2 plots side by side

```python
fig, ax = plt.subplots(1, 2) # 1 row, 2 columns
squares = [x_value**2 for x_value in x]
square_roots = [x_value**0.5 for x_value in x]
ax[0].plot(x, squares)
ax[1].plot(x, square_roots)
plt.show()
```

Or, a 2x2 grid of plots

```python
fig, ax = plt.subplots(2, 2) # 2 rows, 2 columns

squares = [x_value**2 for x_value in x]
cubes = [x_value**3 for x_value in x]
square_roots = [x_value**0.5 for x_value in x]
cube_roots = [x_value**0.33 for x_value in x]

ax[0, 0].plot(x, squares)
ax[0, 1].plot(x, cubes)
ax[1, 0].plot(x, square_roots)
ax[1, 1].plot(x, cube_roots)

plt.show()
```

Sometimes it's not convenient to reference the axes using two indices. In this case, you can use the `flat` attribute of the axes array to iterate over all the axes. `flat` returns a "*1D version*" (technically an iterator) of the array of axes. This is useful e.g. in a `for` loop.

```python
fig, ax = plt.subplots(2, 2) # 2 rows, 2 columns

# Enumerate returns the index of the item being processed as well as the item itself
for i, a in enumerate(ax.flat):
    a.plot(x, [x_values**(i+1) for x_values in x])

plt.show()
```

## How do I customize the plot?

You can change the line style, color, and width using the `linestyle`, `color`, and `linewidth` arguments.

For instance

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5)) # Figure size is 10x5 inches (10 inches wide, 5 inches tall)

ax[0].plot(x, squares, color = 'red', linestyle = '--', linewidth = 2)
ax[1].plot(x, cubes, color = 'green', linestyle = '-.',  linewidth = 2, marker = 'o', markersize = 10)
plt.show()
```

## How do I add labels/titles to the plot?

You can add labels to the x and y axes using the `xlabel` and `ylabel` functions.
You can add a title to the plot using the `title` function (or the `suptitle` function if you want to add a title to the whole figure).

Note that if you're using subplots, the function names are slightly different: `set_xlabel`, `set_ylabel`, and `set_title`.

```python
plt.plot(x, y, color = 'red', linestyle = '--', linewidth = 2)
plt.title ('Square function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

fig, ax = plt.subplots(1, 2, figsize = (10, 5)) 
ax[0].plot(x, cubes, color = 'red', linestyle = 'dotted', linewidth = 2)
ax[0].set_title('Cube function in red')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[1].plot(x, cubes, color = 'blue', marker = 'x')
ax[1].set_title('Cube function in blue')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')

plt.suptitle('Two plots')
plt.show()
```

## Can I have multiple lines in the same plot?

Yes, you can plot multiple lines in the same plot by calling the `plot` function multiple times. Use the `label` argument to give a name to each line, and then call the `legend` function to display the legend.

```python
plt.plot(x, squares, color = 'red', linestyle = 'dotted', linewidth = 2, label = 'y = x^2')
plt.plot(x, [i**0.5 for i in x], color = 'blue', linestyle = 'dashed', linewidth = 2, label = 'y = x^0.5')

plt.legend()
plt.show()
```

You can move the legend around using the `loc` argument.

```python
plt.plot(x, squares, color = 'red', label = 'Square function')
plt.plot(x, [i**0.5 for i in x], color = 'blue', label = 'Square root function')
plt.legend(loc = 'upper right')

plt.show()

plt.plot(x, squares, color = 'red', label = 'Square function')
plt.plot(x, [i**0.5 for i in x], color = 'blue', label = 'Square root function')
plt.legend(loc = 'lower right')

plt.show()
```

# What about other types of plots?

Matplotlib supports many other types of plots, such as bar plots, histograms, images, boxplots etc.

A good way to learn how to create these plots is to look at the [Matplotlib gallery](https://matplotlib.org/stable/gallery/index.html) and the [Matplotlib plot type overview](https://matplotlib.org/stable/plot_types/index.html).

Here are a few examples:

```python
from random import random

vals = [random() for i in range(100)]
plt.hist(vals, bins = 5, color='red')
plt.xlabel('Value') 
plt.ylabel('Frequency')
plt.show()

# A 20x20 matrix of random values
vals = [[random() for i in range(20)] for j in range(20)]
plt.imshow(vals, cmap='Blues')
plt.colorbar()
plt.show()
```
