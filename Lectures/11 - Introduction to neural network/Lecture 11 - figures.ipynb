```python
import matplotlib.pyplot as plt
import numpy as np

# Draw common activation functions
titles = ['Sigmoid', 'Tanh', 'ReLU', 'Leaky ReLU (a = 0.05)']

fig, ax = plt.subplots(1, 4, figsize=(15, 3))
x = np.linspace(-10, 10, 100)
y = 1 / (1 + np.exp(-x))
ax[0].plot(x, y)
y = np.tanh(x)
ax[1].plot(x, y)
y = np.maximum(0, x)
ax[2].plot(x, y)
ax[2].set_ylim(-2, 10)
y = np.maximum(0.05 * x, x)
ax[3].plot(x, y)
ax[3].set_ylim(-2, 10)

for i, a in enumerate(ax):
    a.set_title(titles[i], fontsize=20)

plt.plot()
```
