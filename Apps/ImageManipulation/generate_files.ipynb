```python
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
from skimage.color import rgb2gray
import numpy as np
```

```python
dog_original = imread('dog.jpg')
dog_gray = (rgb2gray(dog_original) * 255).astype(np.uint8)

def convert_to_n_bpp(image:np.ndarray, n_bit:int):
    return (image // (256 // (2 ** n_bit))).astype(np.uint8)

dog_1bit = convert_to_n_bpp(dog_gray, 1)
dog_2bit = convert_to_n_bpp(dog_gray, 2)
dog_4bit = convert_to_n_bpp(dog_gray, 4)
dog_8bit = dog_gray
```

```python
fig, ax = plt.subplots(2, 2, figsize=(7, 7))
ax.flat[0].imshow(dog_gray, cmap='gray')
ax.flat[0].set_title('Original (8-bit)')
ax.flat[0].axis('off')
ax.flat[1].imshow(dog_1bit, cmap='gray', vmin=0, vmax=1)
ax.flat[1].set_title('1-bit')
ax.flat[1].axis('off')
ax.flat[2].imshow(dog_2bit, cmap='gray', vmin=0, vmax=3)
ax.flat[2].set_title('2-bit')
ax.flat[2].axis('off')
ax.flat[3].imshow(dog_4bit, cmap='gray', vmin=0, vmax=15)
ax.flat[3].set_title('4-bit')
ax.flat[3].axis('off')

plt.tight_layout()
```

```python
imsave('dog_1bit.png', dog_1bit)
imsave('dog_2bit.png', dog_2bit)
imsave('dog_4bit.png', dog_4bit)
imsave('dog_8bit.png', dog_8bit)
```
